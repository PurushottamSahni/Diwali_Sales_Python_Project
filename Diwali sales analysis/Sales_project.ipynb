```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.head(10)
```

```python
df.info(10)
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']= df['Amount'].astype('int')
```

```python
df.info(10)
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shadi'})
```

```python
df.describe()
```

```python
df[['Age','Orders', 'Amount']].describe()
```

```python
df.columns
```

```python
ax =sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

From above graphs we can see that most of the customers are female even purchasing power of female is grater than man.

```python
df.columns
```

```python
ax =sns.countplot(x='Age Group', data=df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age =df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

from the above graph we can say that most of the buyers are in the age-group 26-35 and are women.

```python
df.columns
```

```python
sale_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by= 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sale_state, x='State', y='Orders')
```

```python
sale_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sale_state, x='State',y='Amount')
```

From the above graph we can say that UP comes at number 1 in terms of both orders and amount followed by Maharastra and karnataka.

```python
df.columns
```

```python
ax =sns.countplot(x='Marital_Status', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_amount= df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_amount, x='Marital_Status', y='Amount', hue='Gender')
```

from above graph we can  see that most of the buyers are married women and they have high purchasing power.

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,6)})
ax=sns.countplot(data=df, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False )
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(data=sales_occupation, x='Occupation', y='Amount')
```

from above graphs we can see that most buyers are from It, Aviation and Healtcare.

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(25,6)})
ax=sns.countplot(x='Product_Category', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_PC= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(x='Product_Category', y='Amount', data=sales_PC)
```

From above graph we can say that most sold product are from food, clothing and electroins. 

```python
sales_ProductID=df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(x='Product_ID', y='Amount', data=sales_ProductID)
```

Conlusions:

Married women are purchasing more products and also have high purchasing power. Most purchase came from Age Group 26-35. UP is the number one state in terms of orders and purchsing amount followed by Maharastra and Karnataka. Most numbers of items ordered are cloths, food and electronics but in terms of amount food comes at number one.
